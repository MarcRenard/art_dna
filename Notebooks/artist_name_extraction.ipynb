```python
# Load the artists.csv metadata
artists_df = pd.read_csv("raw_data/artists.csv")

# make names lowercase
artists_df["name_clean"] = artists_df["name"].str.lower()
df["artist_name_clean"] = df["artist_name"].str.lower()

# Join on cleaned names
merged_df = df.merge(artists_df, left_on="artist_name_clean", right_on="name_clean", how="left")
print(merged_df.head(5))
```

```python
# Save the result to CSV
df.to_csv("raw_data/image_artist_mapping.csv", index=False)
```
